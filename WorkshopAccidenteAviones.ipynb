Gran parte de la columna Operador tiene la palabra Military eso puede servir para categorizar los vuelos entre los que son civiles y los que pertenecen a las fuerzas militares

```python
aviones['OperadOR'].str.contains('Military')
```

Se crea una nueva columna en la cual cuando la columna Operator tenga la palabra Military la columna tambien lo resaltara.

```python
aviones['CivOMil']=''
conditions=[(aviones['OperadOR'].str.contains("Military"))]
choices=['Military']
aviones['CivOMil']=np.select(conditions,choices)
```

```python
#Los valores que no son Military seran Civiles
aviones['CivOMil'].replace(to_replace='0',value='Civilian',inplace=True)
```

```python
aviones['CivOMil'].head()
```

Todas aquellas columnas en las que necesitamos que dejen hacer operaciones se cambia el '?' por el '0' para poder cambiar su tipo a int.

```python
aviones['all_aboard'].replace(to_replace='?',value='0',inplace=True)
aviones['PASAJEROS A BORDO'].replace(to_replace='?',value='0',inplace=True)
aviones['crew_aboard'].replace(to_replace='?',value='0',inplace=True)
aviones['cantidad de fallecidos'].replace(to_replace='?',value='0',inplace=True)
aviones['passenger_fatalities'].replace(to_replace='?',value='0',inplace=True)
aviones['crew_fatalities'].replace(to_replace='?',value='0',inplace=True)
aviones['ground'].replace(to_replace='?',value='0',inplace=True)
```

```python
aviones['all_aboard']=aviones['all_aboard'].astype('int')
aviones['PASAJEROS A BORDO']=aviones['PASAJEROS A BORDO'].astype('int')
aviones['crew_aboard']=aviones['crew_aboard'].astype('int')
aviones['cantidad de fallecidos']=aviones['cantidad de fallecidos'].astype('int')
aviones['passenger_fatalities']=aviones['passenger_fatalities'].astype('int')
aviones['crew_fatalities']=aviones['crew_fatalities'].astype('int')
aviones['ground']=aviones['ground'].astype('int')
```

```python
#Confirmamos que sean Dtype int64
aviones.info()
```

```python
#Podemos ver que aunque all_aboard es la suma de pasajeros y tripulacion a bordo, esta suma a veces no se ve reflejada
aviones.loc[aviones['all_aboard']==0]
```

```python
#Si la cantidad total de todos a bordo es cero, se suma la tripulacion y pasajeros. Si tambien son 0s la info esta correcta en caso de que no se reemplazara con la suma
aviones['all_aboard'] = np.where((aviones['all_aboard'] == 0), aviones['PASAJEROS A BORDO']+aviones['crew_aboard'],aviones['all_aboard'])
```

Se vuele a utilizar el Where clause para esta vex resolver el hecho de que los pasajeros a bordo o la tripulacion a a bordo son 0 pero no corresponde a la suma de pasajeros totales.

```python
aviones['PASAJEROS A BORDO'] = np.where((aviones['PASAJEROS A BORDO'] == 0), aviones['all_aboard']-aviones['crew_aboard'],aviones['PASAJEROS A BORDO'])
```

```python
aviones['crew_aboard'] = np.where((aviones['crew_aboard'] == 0), aviones['all_aboard']-aviones['PASAJEROS A BORDO'],aviones['crew_aboard'])
```

Se haran exactamente los mismos pasos para laas columnas de cantidad de fallecidos (lo cual es suma de pasajeros fallecidos y tripulacion fallecida).

```python
aviones['cantidad de fallecidos'] = np.where((aviones['cantidad de fallecidos'] == 0), aviones['passenger_fatalities']+aviones['crew_fatalities'],aviones['cantidad de fallecidos'])
```

```python
aviones['passenger_fatalities'] = np.where((aviones['passenger_fatalities'] == 0), aviones['cantidad de fallecidos']-aviones['crew_fatalities'],aviones['passenger_fatalities'])
```

```python
aviones['crew_fatalities'] = np.where((aviones['crew_fatalities'] == 0), aviones['cantidad de fallecidos']-aviones['passenger_fatalities'],aviones['crew_fatalities'])
```

```python
aviones['Categoria_Accidente'] = ''
conditions=[(aviones['Descripcion'].str.contains("engine")),(aviones['Descripcion'].str.contains("pilot")),(aviones['Descripcion'].str.contains("landing")),(aviones['Descripcion'].str.contains("weather")),(aviones['Descripcion'].str.contains("montain")),(aviones['Descripcion'].str.contains("fire")),(aviones['Descripcion'].str.contains("take off")),(aviones['Descripcion'].str.contains("loss")),(aviones['Descripcion'].str.contains("Failure")),(aviones['Descripcion'].str.contains("wing")),(aviones['Descripcion'].str.contains("fog")),(aviones['Descripcion'].str.contains("visual")),(aviones['Descripcion'].str.contains("sea")),(aviones['Descripcion'].str.contains("military")),(aviones['Descripcion'].str.contains("training")),(aviones['Descripcion'].str.contains("maintenance"))]
choices=['engine','pilot','landing','weather','montain','fire', 'take off', 'loss', 'failure', 'wing','visual', 'visual', 'sea','military','training', 'maintenance']
aviones['Categoria_Accidente']=np.select(conditions,choices)
```

```python
#Se usa geo pandas para obtener las coordenadas de la columna Lugar_Accidente, se prueba con un dato individual
import geopandas as gpd
import geopy
from geopandas. tools import geocode
import geopy
geopy.geocoders.options.default_user_agent = "pera"
info=geocode('Fort Myer, Virginia',provider = 'nominatim')
info
```

```python
# Funcion para iterar en la columna Lugar_Accidente
for index, row in aviones.iterrows():
    try:
        print(row['Lugar_Accidente'])
        info = geocode(str(row['Lugar_Accidente']),provider='nominatim')
        aviones.loc[int(index),'Address'] = info['address'].loc[0]
        aviones.loc [int(index),'Lon'] = info[ 'geometry'].loc[0].x
        aviones.loc [int(index),'Lat'] = info[ 'geometry'].loc[0].y
    except TypeError:
        print ('Geocoding information for '+ row['Lugar_Accidente']+" is not found!")
        print ('')
```
